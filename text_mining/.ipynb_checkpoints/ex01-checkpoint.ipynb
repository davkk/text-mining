```python
import string
from pathlib import Path
from typing import cast

import matplotlib.pyplot as plt
import nltk
import seaborn as sns
from nltk.corpus import inaugural, stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download("inaugural")
nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("stopwords")
```

```python
speech: str = cast(str, inaugural.raw("2017-Trump.txt"))
stop_words = stopwords.words("english")

tokens = nltk.word_tokenize(speech.lower())
tokens = set([token for token in tokens if token not in string.punctuation])
tokens_without = [token for token in tokens if token not in stop_words]
```

```python
vec = CountVectorizer()
feat_with = vec.fit_transform(tokens).toarray()
feat_without = vec.fit_transform(tokens_without).toarray()

dot_with = feat_with.T @ feat_with
dot_without = feat_without.T @ feat_without

cos_with = cosine_similarity(feat_with)
cos_without = cosine_similarity(feat_without)
```

```python
plots = [
    (dot_with, "Dot Product with Stopwords"),
    (dot_without, "Dot Product without Stopwords"),
    (cos_with, "Cosine Similarity with Stopwords"),
    (cos_without, "Cosine Similarity without Stopwords"),
]

fig, axs = plt.subplots(figsize=(8, 7), ncols=2, nrows=2)
axs = axs.flatten()

for ax, (matrix, title) in zip(axs, plots):
    sns.heatmap(
        matrix,
        ax=ax,
        xticklabels=False,
        yticklabels=False,
    )
    ax.set_title(title)

fig.tight_layout()
plt.show()
```
